## ให้สร้างฟังก์ชั่นคำนวณเวลาเป็นวินาทีของเวลาใดๆ และ print ออกมาให้สวยงาม

## ให้ สร้าง list ของ เวลา ```['12:30:15','13:41:07',....]``` แล้ววนลูปเรียกฟังก์ชั่นคำนวณเวลาเป็น วินาที

**commit ว่า HW3**

````python
list_time=['12:30:15','13:41:07','12:53:15','00:59:25','11:11:11','16:06:09']
list_time
````

````python
# ฟังก์ชันสำหรับแปลงเวลาเป็นวินาที
def time_to_seconds(time_str):
    # แยกชั่วโมง นาที และวินาทีด้วย split()
    h, m, s = time_str.split(':')
    # แปลงเป็นจำนวนเต็มและคำนวณเวลารวมเป็นวินาที
    seconds = int(h) * 3600 + int(m) * 60 + int(s)
    return seconds

# รายการของเวลาที่ต้องการคำนวณ
time_list = [
    '12:30:15',
    '13:41:07',
    '12:53:15',
    '00:59:25',
    '11:11:11',
    '16:06:09'
]

# วนลูปเรียกฟังก์ชันและพิมพ์ผลลัพธ์
for time_str in time_list:
    seconds = time_to_seconds(time_str)
    print(f"{time_str} = {seconds} วินาที")
````

##ใช้คำสั่ง split()


````python
from datetime import datetime

# สร้าง list เวลา ตัวแรก time1 ตัวที่สอง time2
times = [
    ("12:30:15", "13:41:07"),
    ("12:53:15", "00:59:25"),
    ("16:06:09", "11:11:11")
]

# ฟังก์ชันสำหรับคำนวณความต่างของเวลาในรูปแบบวินาที
def calculate_time_difference(time1_str, time2_str):
    time1_parts = time1_str.split(":")
    time2_parts = time2_str.split(":")

    # แปลงเป็น integer
    time1_hours = int(time1_parts[0])
    time1_minutes = int(time1_parts[1])
    time1_seconds = int(time1_parts[2])

    time2_hours = int(time2_parts[0])
    time2_minutes = int(time2_parts[1])
    time2_seconds = int(time2_parts[2])

    # คำนวณจำนวนวินาทีของแต่ละเวลา
    total_seconds1 = time1_hours * 3600 + time1_minutes * 60 + time1_seconds
    total_seconds2 = time2_hours * 3600 + time2_minutes * 60 + time2_seconds

    # คำนวณความต่างของเวลาในรูปแบบวินาที
    difference_seconds = abs(total_seconds2 - total_seconds1)

    return difference_seconds

# วนลูปคำนวณและแสดงผลลัพธ์
for time1_str, time2_str in times:
    difference_seconds = calculate_time_difference(time1_str, time2_str)
    print(f'ความต่างของเวลา {time2_str} - {time1_str}: {difference_seconds} วินาที')
````

##สร้าง function และเรียกใช้ function ให้มีการวนลูปการคำนวณและแสดงผล
ใช้ tuple ในการเก็บเวลาแล้วแยกออกเป็น time1_str และ time2_str

````python
from datetime import datetime

# สร้าง list เวลา ตัวแรก time1 ตัวที่สอง time2
times = [
    ("12:30:15", "13:41:07"),
    ("12:53:15", "00:59:25"),
    ("16:06:09", "11:11:11")
]

# ฟังก์ชันสำหรับคำนวณความต่างของเวลาในรูปแบบวินาที
def calculate_time_difference(time1_str, time2_str):
    time1 = datetime.strptime(time1_str, "%H:%M:%S")
    time2 = datetime.strptime(time2_str, "%H:%M:%S")
    difference = time2 - time1
    return difference.seconds

# วนลูปคำนวณและแสดงผลลัพธ์
for time1_str, time2_str in times:
    difference_seconds = calculate_time_difference(time1_str, time2_str)
    print(f'ความต่างของเวลา {time2_str} - {time1_str}: {difference_seconds} วินาที')
````
