```python
#สร้างรหัสให้ชื่อ
# สร้าง dictionary เพื่อเก็บรหัสของแต่ละชื่อ
code_dict = {}
code = 1

# สร้าง column ใหม่สำหรับเก็บรหัส
df['name_key'] = ''

# วน loop ผ่านแต่ละแถวใน DataFrame
for index, row in df.iterrows():
  name = row['name']
  if name in code_dict:
    df.at[index, 'name_key'] = code_dict[name]
  else:
    code_dict[name] = code
    df.at[index, 'name_key'] = code
    code += 1

```

```python
# รวม DataFrame ทั้งหมดเข้าด้วยกัน
merged_df = pd.concat(dfs)

```

```python
# จัดกลุ่มข้อมูลตาม 'name_key' และรวมข้อมูล 'cancer_type' เข้าด้วยกัน
result_df = merged_df.groupby('name_key')['cancer_type'].apply(list).reset_index()
result_df
```

```python
# prompt: จาก columns cancer_type แยกออกมาว่า name_key แต่ละอันมีมะเร็งกี่ชนิด  เรียงจากมากไปน้อย

result_df['num_cancer_types'] = result_df['cancer_type'].apply(len)
result_df
```

```python
# นับจำนวนครั้งที่แต่ละชนิดของมะเร็งถูกพูดถึง
cancer_counts = merged_df['cancer_type'].value_counts()

# แสดง 5 ชนิดของมะเร็งที่ถูกพูดถึงมากที่สุด
cancer_counts.head(5)
```

#นำเสนอข้อมูลจุดที่ 3
##ใน 1 โพสต์ ่ พูดถึงมะเร็งมากที่สุดกี่ชนิด และเขาคือใคร พูดถึงว่าอย่างไร

```python
# หาผู้ใช้ที่พูดถึงมะเร็งมากที่สุด
max_cancer_user = result_df.loc[result_df['num_cancer_types'].idxmax()]

# แสดงชื่อผู้ใช้ จำนวนชนิดมะเร็งที่พูดถึง และรายชื่อมะเร็ง
print("ผู้ใช้ที่พูดถึงมะเร็งมากที่สุดคือ:", max_cancer_user['name_key'])
print("จำนวนชนิดมะเร็งที่พูดถึง:", max_cancer_user['num_cancer_types'])
print("รายชื่อมะเร็งที่พูดถึง:", max_cancer_user['cancer_type'])

# หาโพสต์ที่ผู้ใช้คนนี้พูดถึงมะเร็ง
user_posts = merged_df[merged_df['name_key'] == max_cancer_user['name_key']]

# แสดงรายละเอียดของโพสต์
print("\nรายละเอียดของโพสต์:")
for index, row in user_posts.iterrows():
    print(row['comments'])
```
