1. แม่ หรือ พ่อ กี่เรื่อง เรื่องอะไรบ้าง
2.มีคำดังกล่าวในนิทานชีวิตหรือนิทานประจำถิ่นมากกว่ากัน
3.คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่องว่า แม่ หรือ พ่อ มากที่สุด



````
# prompt: วนลูปนิทานที่มีคำว่า แม่ หรือ พ่อ กี่เรื่อง อะไรบ้าง
# มีคำดังกล่าวในนิทานชีวิต หรือนิทานประจำถิ่นมากกว่ากัน
# คนแต่งคนไหน ชอบแต่นิทานที่มีชื่อเรื่อง พ่อ หรือ แม่มากที่สุด

# วนลูปนิทานที่มีคำว่า แม่ หรือ พ่อ กี่เรื่อง อะไรบ้าง
count_mom_dad = 0
list_mom_dad_stories = []
count_life = 0
count_local = 0
creator_count = {}
for index, row in df_folk.iterrows():
  if 'แม่' in row['title'] or 'พ่อ' in row['title']:
    count_mom_dad += 1
    list_mom_dad_stories.append(row['title'])
    if row['type'] == 'นิทานชีวิต':
      count_life += 1
    elif row['type'] == 'นิทานประจำถิ่น':
      count_local += 1
    if row['creator'] in creator_count:
      creator_count[row['creator']] += 1
    else:
      creator_count[row['creator']] = 1

print(f"มีนิทานที่มีคำว่า 'แม่' หรือ 'พ่อ' ทั้งหมด {count_mom_dad} เรื่อง")
print("รายชื่อนิทาน:")
for story in list_mom_dad_stories:
  print(story)

# มีคำดังกล่าวในนิทานชีวิต หรือนิทานประจำถิ่นมากกว่ากัน
if count_life > count_local:
  print("มีคำว่า 'แม่' หรือ 'พ่อ' ในนิทานชีวิตมากกว่า")
elif count_local > count_life:
  print("มีคำว่า 'แม่' หรือ 'พ่อ' ในนิทานประจำถิ่นมากกว่า")
else:
  print("มีคำว่า 'แม่' หรือ 'พ่อ' ในนิทานชีวิตและนิทานประจำถิ่นเท่ากัน")

# คนแต่งคนไหน ชอบแต่นิทานที่มีชื่อเรื่อง พ่อ หรือ แม่มากที่สุด
max_count = 0
most_frequent_creator = ""
for creator, count in creator_count.items():
  if count > max_count:
    max_count = count
    most_frequent_creator = creator

print(f"คนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่อง 'พ่อ' หรือ 'แม่' มากที่สุดคือ {most_frequent_creator} ({max_count} เรื่อง)")
````

````
import pandas as pd

# ใช้เกณฑ์ จำนวนภาพยนต์ที่กำกับ rating เฉลี่ย และ จำนวนโหวตรวม ในการพิจารณาการเป็นผู้กำกับชั้นนำ
director_stats = movie_df.groupby('Director').agg({
    'Name': 'count',          # จำนวนภาพยนตร์ที่กำกับ
    'Rating': 'mean',         # คะแนนเฉลี่ย
    'Votes': 'sum'            # จำนวนโหวตรวม
}).rename(columns={'Name': 'Films_Directed', 'Rating': 'Avg_Rating', 'Votes': 'Total_Votes'}).reset_index()



````

````
director_stats
````

````
# Convert 'Duration' column back to string type before applying string methods
movie_df['Duration'] = movie_df['Duration'].astype(str).str.replace(' min', '').astype(int)
````

````
# จัดอันดับผู้กำกับตามเกณฑ์
director_stats['Rank'] = director_stats[['Films_Directed','Avg_Rating', 'Total_Votes']].rank(method='max', ascending=False).mean(axis=1)

# เรียงลำดับผู้กำกับตามอันดับ
top_directors = director_stats.sort_values(by='Rank').head(3)
print(top_directors)
````

````
top_directors_list = top_directors['Director'].tolist()

for director in top_directors_list:
    director_info = movie_df[movie_df['Director'] == director]
    print(f"Director: {director}")
    print(f"ทำไมถึงเป็นผู้กำกับชั้นนำ:")
    print(f"- จำนวนภาพยนตร์ที่กำกับ: {director_info['Name'].count()}")
    print(f"- คะแนนเฉลี่ย: {director_info['Rating'].mean():.2f}")
    print(f"- จำนวนโหวตรวม: {director_info['Votes'].sum()}")
    print(f"พวกเขาเป็นอย่างไร:")
    print(f"{director} เป็นที่รู้จักในด้านสไตล์ที่เป็นเอกลักษณ์และมีส่วนสำคัญต่อวงการภาพยนตร์อินเดีย ภาพยนตร์ของพวกเขาได้รับการตอบรับอย่างดีจากผู้ชมและนักวิจารณ์")
    print()
````
